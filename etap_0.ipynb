```python
import numpy as np
import pandas as pd
import igraph as ig

%%latex
```

```python
df = pd.read_csv('Dane/SMALL/data10.txt', skiprows=2, sep='\s+', names=['A','B','C','D','E','F','G','H','I','J'])
```

```python
df
```

```python
ar = df.values
ar
```

```python
graph = ig.Graph(10)
graph.es["weight"] = 1.0

for i in range(10):
    for j in range(10):
        if ar[i,j] != -1:
            graph[i,j]=ar[i,j]
```

```python
graph.es["weight"]
```

```python
plot(graph, edge_label=graph.es["weight"], vertex_label=['A','B','C','D','E','F','G','H','I','J'])
```
