```python
unique_czy_hybrydowy = [True,False]
unique_wersja_ls = [0,1,-1]
unique_nr_selekcji = [0,1,2,3]
unique_precyzja_ruletki = [0,1,2,-1]
unique_ilosc_uczstnikow_turnieju = [2,10,-1]
unique_metoda_selekcji_rankingowej = [0,1,-1]
unique_metoda_krzyzowania = [0,1,2]
```

```python
kombinacje = {}
for a in unique_czy_hybrydowy:
    for b in unique_wersja_ls:
        for c in unique_nr_selekcji:
            for d in unique_precyzja_ruletki:
                for e in unique_ilosc_uczstnikow_turnieju:
                    for f in unique_metoda_selekcji_rankingowej:
                        for g in unique_metoda_krzyzowania:
                            kombinacje[a,b,c,d,e,f,g] = 0
```

```python
for list_max in df_max:
    for index, row in list_max.iterrows():
        kombinacje[row['czy_hybrydowy'],row['wersja_ls'],row['nr_selekcji'],row['precyzja_ruletki'],row['ilosc_uczestnikow_turnieju'],row['metoda_selekcji_rankingowej'],row['metoda_krzyzowania']] += 1
```

```python
for key, values in kombinacje.items():
    if values == 2:
        print(key)
```
